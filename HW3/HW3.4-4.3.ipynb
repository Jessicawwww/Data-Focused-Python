## Expressing conditional logic as array options

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
```

```python
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
cond = np.array([True, False, True, True, False])
```

```python
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
# not be very fast for large arrays
# not work with multidimensional arrays.
```

```python
result
```

```python
result = np.where(cond, xarr, yarr)
#The second and third arguments to np.where don’t need to be arrays; 
# one or both of them can be scalars
```

```python
result
```

```python
arr = np.random.randn(4,4)
```

```python
arr
```

```python
arr>0
```

```python
np.where(arr>0, 2, -2)
```

```python
np.where(arr>0, 2, arr)
```

## 4.7 Example: Random Walks

```python
import random
```

```python
position = 0
walk = [position]
steps=1000
```

```python
for i in range(steps):
    step=1 if random.randint(0,1) else -1
    position +=step
    walk.append(position)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(walk[:100])
```

```python
nsteps=1000
```

```python
draws = np.random.randint(0,2,size = nsteps)
```

```python
steps = np.where(draws>0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
(np.abs(walk)>=10).argmax()
```

## Simulating many random walks at once

```python
nwalks=5000
nsteps=1000
```

```python
draws = np.random.randint(0,2,size = (nwalks, nsteps)) #0 or 1
```

```python
steps = np.where(draws>0, 1, -1)
```

```python
walks = steps.cumsum(1)
```

```python
walks
```

```python
walks.max()
```

```python
walks.min()
```

```python
hits30 = (np.abs(walks)>=30).any(1)
```

```python
hits30
```

```python
hits30.sum() #number that hit 30 or -30
```

```python
crossing_times = (np.abs(walks[hits30])>=30).argmax(1)
```

```python
crossing_times.mean()
```
